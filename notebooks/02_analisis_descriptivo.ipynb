```python
!wget -O Housing.xlsx "https://github.com/FranciscoRamirezRuiz/Inferencia2024/raw/main/data/Housing.xlsx"

import pandas as pd

df = pd.read_excel("Housing.xlsx")
df.head()
```

```python

df.describe().T
```

```python
import matplotlib.pyplot as plt

columnas = ['price', 'sqft_living', 'bedrooms']

plt.figure(figsize=(16, 4))

for i, col in enumerate(columnas):
    plt.subplot(1, len(columnas), i+1)
    df[col].hist(bins=30)
    plt.title(f'Histograma de {col}')

plt.tight_layout()
plt.show()
```

```python
import seaborn as sns

for col in columnas:
    plt.figure()
    sns.boxplot(x=df[col])
    plt.title(f"Boxplot de {col}")
    plt.show()
```

```python
corr = df.corr(numeric_only=True)

plt.figure(figsize=(14, 10))
sns.heatmap(corr, annot=True, fmt=".2f", cmap="coolwarm", square=True)
plt.title("Mapa de calor - correlación entre variables numéricas")
plt.show()
```

```python
from scipy.stats import shapiro

variables = ['price', 'sqft_living', 'bathrooms']

for var in variables:
    stat, p = shapiro(df[var])
    print(f"Variable: {var} - Estadístico: {stat:.4f}, p-valor: {p:.4f}")
    if p < 0.05:
        print(" No se puede asumir normalidad (p < 0.05)")
    else:
        print(" Se puede asumir normalidad (p ≥ 0.05)")
```
